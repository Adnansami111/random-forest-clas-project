```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,r2_score,precision_score
```

```python
df=pd.read_excel("random_forest_classification_dataset.xlsx")
```

```python
columns=['Age', 'Income', 'Credit_Score', 'Loan_Amount', 'Savings', 'Dependents','Loan_Approved']
for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"{col}: {outliers.shape[0]} outliers")
```

```python
for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
X = df.drop("Loan_Approved",axis=1)
```

```python
Y = df["Loan_Approved"]
```

```python
scaler = StandardScaler()
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state=42,test_size=0.2)
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
model = RandomForestClassifier(n_estimators = 12,max_depth=6,random_state=42)
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
print(accuracy_score(Y_test,Y_pred))
```
